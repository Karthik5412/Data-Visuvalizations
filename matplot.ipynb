```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Data Frame

```python
x = np.linspace(0,15,12).round(2)
```

```python
y = x ** 5
```

```python
x
```

```python
y = y.round(2)
```

```python
y
```

# Creating basic plot

```python
plt.plot(x,y,color = 'darkgreen',lw = 3, ls= '-.')
plt.title('Basic Plot')
plt.xlabel('X values')
plt.ylabel('Y values')
```

# Subplots

```python
plt.subplot(2,2,1)
plt.plot(x,y, color='purple',lw=3)
plt.subplot(2,2,2)
plt.plot(x,y * -1, color='teal',lw=3)
plt.subplot(2,2,3)
plt.plot(x,y * -10, color='black',lw=3)
plt.subplot(2,2,4)
plt.plot(x,y * 100, color='maroon',lw=3)
```

# Object Oriented Way to create subplots

```python
# creating object for figure class
fig = plt.figure()
# Accessing the axis function which is present in figure class
ax1 = fig.add_axes([0.3,0.3,.5,.5])
ax1.plot(x,y)
ax2 = fig.add_axes([1,0.3,.5,.5])
ax2.plot(y,x)
```

```python
# using the same example to demonstrate the figure effecrs like depth and figsize
fig2 = plt.figure(figsize = (10, 6), dpi = 120)
# Accessing the axis function which is present in figure class
axa = fig2.add_axes([0.3,0.3,.5,.5])
axa.plot(x,y)
axb = fig2.add_axes([1,0.3,.5,.5])
axb.plot(y,x)
```

# Types of plots

```python
# Scatter plot
plt.scatter(x,y,color = 'green')
```

```python
# Histogram
plt.hist(x,color='yellow')
```

```python
# Box Plot
plt.boxplot(y)
```

# Saving the plot

```python
plt.subplot(2,2,1)
plt.plot(x,y, color='purple',lw=3,ls='--')
plt.subplot(2,2,2)
plt.plot(x,y * -1, color='seagreen',lw=3,ls=':')
plt.subplot(2,2,3)
plt.plot(x,y * -10, color='black',lw=3,ls='-.')
plt.subplot(2,2,4)
plt.plot(x,y * 100, color='maroon',lw=3)

# Save syntax
plt.savefig('mat.png')
```
