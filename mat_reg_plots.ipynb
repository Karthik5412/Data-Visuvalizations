```python
import seaborn as sn
import scipy as sc
import pandas as pd
```

```python
tips=sn.load_dataset('tips')
```

# Matrix plots

## Heat Map

```python
cr_tips = tips.select_dtypes(include=['number']).corr()
```

```python
sn.heatmap(cr_tips)
```

## Cluster Map

```python
# %pip install scipy

sn.clustermap(data=cr_tips)
```

## Pivot Table Heat Map

```python
# creating pivot table
pvt = pd.pivot_table(data=tips,columns='size',index='day',values='total_bill')
```

```python
# We are building the heat map on it
sn.heatmap(pvt)
```

# Regression Plot

```python
# For creating this we use lmplot
sn.lmplot(x='total_bill', y= 'tip', data = tips, palette='dark')
```
