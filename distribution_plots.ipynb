```python
import seaborn as sn
import matplotlib.pyplot as plt
import pandas as pd
```

# Data Frames

```python
coffee = pd.read_csv("D:\\Py_prgms\\panda\\coffee.csv")
titanic = sn.load_dataset('titanic')
flights = sn.load_dataset('flights')
tips = sn.load_dataset('tips')
irs = sn.load_dataset('iris')
png = sn.load_dataset('penguins')
```

# Dist plot

```python
sn.displot(flights,palette= 'rainbow')
```

```python
# Here we are increasing the bin size
sn.displot(data = flights, bins = 20, palette= 'dark')
```

# Joint plot

```python
sn.jointplot(flights,kind= 'scatter', palette='mako')
```

```python
# hist
sn.jointplot(flights,kind= 'hist', palette='mako')
```

```python
# kde
sn.jointplot(titanic,kind= 'kde', palette='rainbow')
```

```python
#reg
sn.jointplot(data=tips,x='total_bill',y='tip',kind= 'hex', palette='mako')
```

```python
#reg
sn.jointplot(data=tips,x='total_bill',y='tip',kind= 'reg', palette='mako')
```

```python
#reg
sn.jointplot(data=tips,x='total_bill',y='tip',kind= 'resid', palette='mako')
```

# Pair plots

```python
sn.pairplot(data= tips,hue= 'smoker', palette='rainbow')
```
