```python
import seaborn as sn
import numpy as np
```

# Data Frames

```python
png = sn.load_dataset('penguins')
tips = sn.load_dataset('tips')
```

# Count plot

```python
sn.countplot(x=tips['sex'],hue = tips['smoker'],palette = 'crest')
```

# Bar plot

```python
sn.barplot(x=png['bill_length_mm'],y=png['bill_depth_mm'],estimator=np.std)
```

# Box Plot

```python
sn.boxplot(y='total_bill', x='tip',data= tips)
```

# Violin plot

```python
sn.violinplot(y='total_bill', x='tip',data= tips)
```

# Stripplot

```python
sn.stripplot(y='total_bill', x='tip',data= tips)
```

# Swarm Plot

```python
sn.swarmplot(y='total_bill', x='tip',data= tips)
```
